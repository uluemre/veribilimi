```python
import pandas as pd
df = pd.read_csv("country.csv")
df.head()
```

```python
df_sorted = df.sort_values(by="Population", ascending=False)
df_sorted.head(20)
```

```python
df_gdp_sorted = df.sort_values(by="GDP ($ per capita)", ascending=True)
df_gdp_sorted.head(10)
```

```python
df_pop_over_10m = df[df["Population"] > 10000000]
df_pop_over_10m.head()
```

```python
df_literacy_sorted = df.sort_values(by="Literacy (%)", ascending=False)
df_literacy_sorted.head(5)
```

```python
df_gdp_over_10k = df[df["GDP ($ per capita)"] > 10000]
df_gdp_over_10k.head()
```

```python
df_density_sorted = df.sort_values(by="Pop. Density (per sq. mi.)", ascending=False)
df_density_sorted.head(10)
```
