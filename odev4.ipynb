NumPy Bölümü

```python
import numpy as np

matris = np.random.randint(0, 101, (5, 5))
print("Matris:\n", matris)

print("Ortalama:", np.mean(matris))
print("Standart Sapma:", np.std(matris))
print("Varyans:", np.var(matris))

print("En büyük değer:", np.max(matris))
print("En küçük değer:", np.min(matris))

print("Köşegen toplamı:", np.trace(matris))
```

```python
puanlar = np.random.normal(70, 15, 1000)
puanlar = np.clip(puanlar, 0, 101)

print("Ortalama:", np.mean(puanlar))
print("Medyan:", np.median(puanlar))
print("Standart Sapma:", np.std(puanlar))

print("50’den düşük alan öğrenci sayısı:", np.sum(puanlar < 50))
```

Pandas Bölümü

```python
import pandas as pd

data = {
    "Öğrenci": ["Ali", "Ayşe", "Mehmet", "Zeynep", "Ahmet"],
    "Yaş": [20, 21, 19, 22, 20],
    "Bölüm": ["Bilgisayar", "Fizik", "Kimya", "Bilgisayar", "Fizik"],
    "Matematik": [70, 60, 80, 90, 55],
    "Fizik": [65, 75, 70, 85, 60],
    "Kimya": [80, 85, 65, 95, 70]
}

df = pd.DataFrame(data)
print(df)
```

```python
print("Her dersin ortalaması:\n", df[["Matematik","Fizik","Kimya"]].mean())
```

```python
max_math = df.loc[df["Matematik"].idxmax()]
print("En yüksek matematik notu alan öğrenci:\n", max_math)
```

```python
df["Ortalama"] = df[["Matematik","Fizik","Kimya"]].mean(axis=1)
print(df)
```

```python
print("Bölümlere göre ortalama başarı:\n", df.groupby("Bölüm")[["Matematik","Fizik","Kimya","Ortalama"]].mean())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Her dersin histogramı
df[["Matematik","Fizik","Kimya"]].hist(bins=10, figsize=(10,5))
plt.suptitle("Derslerin Not Dağılımları")
plt.show()

# Bölümlere göre ortalama başarı bar grafiği
df.groupby("Bölüm")[["Ortalama"]].mean().plot(kind="bar", legend=False)
plt.title("Bölümlere Göre Ortalama Başarı")
plt.ylabel("Ortalama Not")
plt.show()
```
